```python
# 2

# Jupyter Note: Remember to run all cells in order to use all variables defined in this file
import numpy as np
import matplotlib.pyplot as plt

# Read in data from Excel file
ws = np.loadtxt("weatherstats_vancouver_daily.csv", delimiter=",", usecols=(17), skiprows=(1))
print(ws)
length=len(ws)

# Adding all elements of the Windspeed coloum 
i=0
sum=0
for i in range(0,length-1):
    sum += ws[i]
    
# Using the length and sum of elements of ws, calculate the mean wind speed
mean1=sum/length
print ("Mean of the average daily wind speeds %.3f km/h" % (mean1)) 





```

```python
#3

# Just a histogram plot
plt.hist(ws, bins='auto')
plt.show
```

```python
#4

import math  # Library containing important math functions

# subtracting each element of ws with the mean and squaring to get standard deviation
sum1=0
s1=ws-mean1
s2= s1**2
#Adding the squared differences 
for j in range(0,length-1):
    sum1 += s2[j]

# length is unchanged, so square root of quotient gives std
std1=math.sqrt(sum1/length)
# print(std)

# plotting a histogram with various settings and titles
n, bins, patches=plt.hist(ws, 50)
plt.setp(patches[15], 'facecolor', 'r')
plt.text(20,800, r'$\bar{x}=$%5.2f $\sigma=$%5.2f' % (mean1, std1))
plt.axis([0, 50, 0, 1250])
plt.ylabel('Frequency of Bin')
plt.xlabel('Average Wind Speeds')
plt.title('Histogram of Average Wind Speeds in Vancouver')

plt.grid(True)
plt.show()
```
