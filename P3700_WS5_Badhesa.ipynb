```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

N=200
x=np.linspace(0,3,N)
p=np.linspace(-np.pi,3*np.pi,N)
y=[0]*N
z=[0]*N
q=[0]*N
d=[0]*N
g=[0]*N
s=[0]*N

c=[0]*N
u=[0]*N


h=[0]*N
r=[0]*N

e=[0]*N
f=[0]*N

sum=0

i=0
for i in range(0,N):
    #DFT
    h[i]= 2-(np.cos(np.pi*x[i]/2))+3*(np.cos(np.pi*x[i]))+3*(np.cos(3*np.pi*x[i]/2))
    r[i]= np.sin(np.pi*x[i]/2)-3*np.sin(np.pi*x[i])-3*np.sin(3*np.pi*x[i]/2)
    #DTFT
    d[i]= 2-(np.cos(p[i]))+3*(np.cos(2*p[i]))+3*(np.cos(3*p[i]))
    g[i]= np.sin(p[i])-3*np.sin(2*p[i])-3*np.sin(3*p[i])
    
    
    c[i]=d[i]**2
    u[i]=g[i]**2
    
    g[i]=complex(h[i],r[i])
    
    y[i]=h[i]**2
    z[i]=r[i]**2
    e[i]=np.sqrt(y[i]+z[i])
    s[i]=np.sqrt(c[i]+u[i])
    
plt.figure(figsize=(12,8))
plt.plot(x, e,label = "DFT")
plt.plot(p, s,label = "DTFT")
#plt.ylim(-2, 2)
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("H(z)")
plt.grid(True)
plt.legend(loc= 'upper right')
plt.show()
```
