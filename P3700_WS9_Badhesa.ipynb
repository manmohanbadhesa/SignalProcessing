```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt

import time
time_start = time.process_time()
#run your code

ascent = scipy.misc.ascent()
plt.gray()
plt.imshow(ascent)
plt.show()

# Histogram of the pixel values

plt.hist(ascent.ravel(), bins='auto') # Quick histogram with settings on 'auto'
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Pixel Values of the Ascent Image Histogram')
plt.show()

time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```

```python
# b) increased brightness

wa,ha=512,512 #w for width and h for height
b=[[0 for x in range(wa)] for y in range(ha)]

# The brightness is increased by increasing the pixel values while ensuring that the pixel values don't exceed 255 and the 
# ones increasing are made equal to 255
for i in range(0,wa):
    j=0
    for j in range(0,ha):
        b[i][j]= ascent[i][j]+100
        if b[i][j]>255:
            b[i][j]=255
        #print(b[i][j],ascent[i][j])

plt.gray()
plt.imshow(b)
plt.show()
```

```python
# c) decreased brightness

c=[[0 for x in range(wa)] for y in range(ha)]


# Brightness decreased by decreasing pixel values but ensuring that pixel values are not negative and negative ones are 
# made 0
for i in range(0,wa):
    j=0
    for j in range(0,ha):
        c[i][j]= ascent[i][j]-100
        if c[i][j]<0:
            c[i][j]=0
        #print(c[i][j],ascent[i][j])

plt.gray()
plt.imshow(c)
plt.show()
```

```python
# d) slightly increased contrast at DN75

d=[[0 for x in range(wa)] for y in range(ha)]


# slope depends on start and stop, so can adjust to change slope
start=40
stop=110
slope = int(255/(stop-start))
intercept=(255-slope*stop)

# Using the equation of the slope of the line, the slope is increased around pixel value 75, and still ensured that pixel 
# values are between 0 and 255
for i in range(0,wa):
    j=0
    for j in range(0,ha):
        if ascent[i][j] <start:
                d[i][j]=0
        if ascent[i][j]>=start and ascent[i][j]<=stop:
            d[i][j]= ascent[i][j]*slope+intercept
        if d[i][j]>255 or ascent[i][j]>stop:
            d[i][j]=255
        if d[i][j] <0:
            d[i][j]=0
            

        #print(d[i][j],ascent[i][j])


plt.gray()
plt.imshow(d)
plt.show()
```

```python
# e) Greatly increased Contrast at DN 150

e=[[0 for x in range(wa)] for y in range(ha)]

# slope depends on start and stop, so can adjust to change slope
start=140
stop=160
slope = int(255/(stop-start))
intercept=(255-slope*stop)

# Using the equation of the slope of the line, the slope is increased  signficantly around pixel value 150, and still 
# ensured that pixel values are between 0 and 255
for i in range(0,wa):
    j=0
    for j in range(0,ha):
        if ascent[i][j]< start:
            e[i][j]=0
        if ascent[i][j]>=start and ascent[i][j]<=stop:
            e[i][j]= ascent[i][j]*slope+intercept
        if e[i][j]>255 or ascent[i][j]> stop:
            e[i][j]=255
        if e[i][j]<0:
            e[i][j]=0

        
        #print(e[i][j],ascent[i][j])

plt.gray()
plt.imshow(e)
plt.show()
```

```python
# f) Contrast change at DN 75 and 150

f=[[0 for x in range(wa)] for y in range(ha)]

# same as d and e except changing slope at two pixel values of 75 and 150
start=70
stop=90
start2=220
stop2=240
slope = int(255/(stop-start))
intercept=int((125)-slope*stop)
intercept2=int((255)-slope*stop2)
#print(intercept,intercept2,slope)

for i in range(0,wa):
    j=0
    for j in range(0,ha):
        if ascent[i][j] < start:
            f[i][j]=0
        if ascent[i][j]>=start and ascent[i][j]<=stop:
            f[i][j]= ascent[i][j]*slope+intercept
        if ascent[i][j]>=start2 and ascent[i][j]<=stop2:
            f[i][j]= ascent[i][j]*slope+intercept2       
        if ascent[i][j]>=stop and ascent[i][j]<=start2:
            f[i][j]= stop*slope+intercept        
        if f[i][j]>255 or ascent[i][j]>stop2:
            f[i][j]=255
        if f[i][j]<0:
            f[i][j]=0
        
        #print(f[i][j],ascent[i][j])

plt.gray()
plt.imshow(f)
plt.show()
```

```python
import matplotlib.image as mpimg
img1=mpimg.imread("clock1.png")
img2=mpimg.imread("clock2.png")

# converting the rgb to gray scale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

gray1=np.floor(rgb2gray(img1)*8+0.5)
plt.imshow(gray1,cmap=plt.get_cmap('gray'))
plt.show()
gray2=np.floor(rgb2gray(img2)*8+0.5)
plt.imshow(gray2,cmap=plt.get_cmap('gray'))
plt.show()

# subtracting the two clock images
clocksub=gray1-gray2

plt.imshow(clocksub)
plt.show()

# this writes out the 2D list imgsub into a png file
mpimg.imsave("clocksub.png",clocksub)
```
