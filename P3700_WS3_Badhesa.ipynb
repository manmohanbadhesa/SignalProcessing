```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# generating square wave
N=500
square=[0]*N
 
i=200
for i in range(i,300):
    square[i] += 1

    # adding uniform noise
noise= np.random.uniform(-0.6,0.6,N)
nS=[0]*N                               # noisy square wave
k=0
for k in range(0,N):
    nS[k]=square[k]+noise[k]

# Fourier transform the noisy square wave
fft1 = np.fft.fft(nS) #xsamp is your stem plot data
N1 = len(nS)
Tsamp1=1/N
plt.plot(nS)
plt.show()

#plt.figure(figsize=(12,4))
plt.xlabel("Sample Value")
plt.ylabel("Amplitude")
plt.title("Noisy Square Wave ")
plt.plot(np.abs(fft1)/N1)
plt.show()

# Applying averging filter to nosiy wave
pointAvgNumber=11
r=N/pointAvgNumber
ratio=int(r)
Filter1=[0]*N
Filter2=[0]*N
i=0
# going through the noisy wave list and averging them the first 11 points and then removing the first element and
# adding the next element, averging 11 points everytime and storing to new array called filter
for i in range(0,N-11):
    j=0
    for j in range(0,pointAvgNumber):
        #print(i,j+i)
        Filter1[i] += nS[j+i]
        Filter2[i]= Filter1[i]/11


    
fft2 = np.fft.fft(Filter2) #xsamp is your stem plot data
N2 = len(Filter2)
Tsamp2=1/ratio

plt.plot(Filter2)
plt.show()

#plt.figure(figsize=(12,4))
plt.xlabel("Sample Value")
plt.ylabel("Amplitude")
plt.title("Filtered Square Wave ")
plt.plot(np.abs(fft2)/N2)
plt.show()
```
