# 1 a)

```python
import numpy as np
import matplotlib.pyplot as plt


# Loading the data from the "arb_freq_resp.csv" file to get fake frequency data to plot

arbfreq = np.loadtxt("arb_freq_resp.csv", delimiter=",", usecols=(0))
r1=len(arbfreq)

axi=np.linspace(0,np.pi,r1)

plt.figure(figsize=(12,8))
plt.plot(axi,arbfreq)
plt.xlabel('\u03A9')
plt.ylabel('H(\u03A9)')
plt.title("Arbritrary Frequency Response")
plt.show()
```

# 1 b)

```python
# IDFT to the above data to switch from frequency domain to time domain


b1IDFT =np.fft.ifft(arbfreq)
R1=len(b1IDFT)

# just an array to plot against
fs1=2*np.pi
Tsamp1=1/fs1
F1 = np.linspace(0, 1 / Tsamp1, R1)

plt.figure(figsize=(12,8))
plt.plot(F1,b1IDFT)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.title("IDFT")
plt.show()
```

# 1 c)

```python
# shift and truncate IDFT to get filter kernel

M=30
M1=int(M/2)
print(M1)
fkern1=[0.]*M


# Only using first 15 points and last 15 points of the time domain data for our filter kernel
for i in range(0,M1):
    fkern1[i]=b1IDFT[R1-M1+i]
    fkern1[i+M1]=b1IDFT[i]

plt.figure(figsize=(12,8))
plt.plot(fkern1)
plt.xlabel('Kernel Number')
plt.ylabel('Amplitude')
plt.title("Filter Kernal")
plt.show()
```

# 1 d)

```python
# Applying dft to the kernel gives us somewhat the origianl freq response but not entirely the same. The major shape exists

Ftr1=np.fft.fft(fkern1)

axi1=np.linspace(0,np.pi,M)

plt.figure(figsize=(12,8))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Fourier Transform Plot ")
plt.plot(axi1,np.abs(Ftr1)/M)
plt.show()
```

# 1 e)

```python
# Recalling hamming, hanning and blackman window functions from Worksheet 6 to apply DTFT padded with zeroes

import time
time_start = time.process_time()
#run your code


def han(N):
    windw=[0.]*N
    for i in range(-int(N/2),int(N/2)):
        windw[i+int(N/2)]=0.5+0.5*np.cos(2*np.pi*i/(N-1))
    return windw

def ham(N):
    windw=[0.]*N
    for i in range(int(-N/2),int(N/2)):
        windw[i+int(N/2)]=0.54+0.46*np.cos(2*np.pi*i/(N-1))
    return windw
def BM(N):
    windw=[0.]*N
    for i in range(int(-N/2),int(N/2)):
        windw[i+int(N/2)]=0.5+0.5*np.cos(2*np.pi*i/(N-1))+0.08*np.cos(4*np.pi*i/(N-1))
    return windw

time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```

```python
# The hamming window was applied to the time domain filter kernel, the hamming window smoothed out the kernel to
# resemble the original freq response more while stil maintaining the sharp corners

ham1=ham(M)

fkern2=[0.]*M

for i in range(0,M):
    fkern2[i]=ham1[i]*fkern1[i]

plt.figure(figsize=(12,8))
plt.plot(fkern2)
plt.xlabel('Kernel Number')
plt.ylabel('Amplitude')
plt.title("Filter Kernal")
plt.show()

Ftr2=np.fft.fft(fkern2)

plt.figure(figsize=(12,8))
plt.plot(np.abs(Ftr2))
plt.xlabel('Kernel Number')
plt.ylabel('Amplitude')
plt.title("Filter Kernal")
plt.show()
```

```python
# Using the dft of detected and desired, we find the dft of the kernel by dividing the dft of desired by dft of the 
# detected. Then take the IDFT to find the kernel

dftdet=np.fft.fft(detected)
dftdes=np.fft.fft(desired)

dftkern=dftdes/dftdet

kern=np.fft.ifft(dftkern)

plt.figure(figsize=(12,8))
plt.plot(kern)
plt.xlabel('Kernel Number')
plt.ylabel('Amplitude')
plt.title("Deconvolved Kernel")
plt.show()
```

# 2 e)

```python
# convolve function from previous lab

def myconv(list1,list2):
    len1=len(list1)
    len2=len(list2)
    
    # Below we assign the longer list to x and shorter list to h to ensure 
    # the elements compiling efficiency
    if len1> len2:
        h=list2
        x=list1
    else:
        h=list1
        x=list2
    len3=len1+len2-1
    #print(len3)
    y=[0]*len3
    i=0
    # below we loop through summing up our elements as h moves through x 
    for i in range(0,len3):
        j=0
        for j in range(0,len(h)):
            if i-j >=0:
                if i-j < len(x):
                    y[i]+=x[i-j]*h[j]
                    #print(y[i], x[i-j], h[j], i,j,i-j)
    return y

# convolving the kernel found in 2d) and the imported gamma data, gammaT contains the data that have seperate impulses 
# of gamma rays

gammaT=myconv(kern,gamma)

plt.figure(figsize=(12,8))
plt.plot(gammaT)
plt.xlabel('Pulse Number')
plt.ylabel('Pulse Amplitude')
plt.title("Convolved Gamma data")
plt.show()
```

```python
# lowSync function is a function from previous lab which generates a N number of kernels

def lowSync(L5):
    lsyn=[0]*L5
    i=0
    k=1/100000000
    sum=0
    L8=int(L5/2)
    for i in range(-L8,L8):
        if i==0:
            lsyn[i+L8]=np.sin(i+k)/(i+k)
            #print(lsyn[i+L8],i+1,np.sin(i))
        else:
            lsyn[i+L8]=np.sin(i)/(i)
        sum +=lsyn[i+L8]
        #print(i)
    j=0
    for j in range(0,L5):
        lsyn[j]= lsyn[j]/sum
        #print(lsyn[j])
    return lsyn

# 41 point sinc kernels
K=41
sinckern=lowSync(K)

# dft of the sinc which needs to be 256 points 

L=256

sinc=[0.]*L
sinckern=lowSync(K)


for i in range(0,K):
    sinc[i]=sinckern[i]
    
dftsinc=np.fft.fft(sinc)
```

```python
time_start = time.process_time()
#run your code

# Number of divisions of the quick brown fox data
a1=len(A1)
segN=int(a1/128)

# The last segment of data isnt exactly 128 points so is dealt seperately
u=128
remainderPoints=a1-(segN*u)
print(remainderPoints)

# Initialling segN number of segments containing 256 elements each
segments=[0]*(segN+1)
for i in range(0,segN+1):
    segments[i]=[0]*L
#print(segments[1][0])
#print(segments[0][0])


# the actual separation of the elements into seperate columns, only of the first 128 points of each column, rest zeros
h=-128
for i in range(0,segN):
    j=0
    h+=128
    for j in range(0,u):
        segments[i][j]=A1[j+h]
        #print(i,j,segments[i][j],A1[j+h])
for i in range(0,remainderPoints):
    segments[segN][i]=A1[segN*u+i]

    
# dft of each segments stored in similar way as segments

dftsegments=[0]*(segN+1)
for i in range(0,segN):
    dftsegments[i]=[0]*L
for i in range(0,segN+1):
    dftsegments[i]=np.fft.fft(segments[i])


    
time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```

```python
time_start = time.process_time()
#run your code

# multiplying each segment with the sinc kernels
product1=[0]*(segN+1)
for i in range(0,segN+1):
    product1[i]=[0]*L

u=256

for i in range(0,segN+1):
    j=0
    for j in range(0,u):
        #print(i,j)
        product1[i][j]=dftsegments[i][j]*dftsinc[j]
        
time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```
