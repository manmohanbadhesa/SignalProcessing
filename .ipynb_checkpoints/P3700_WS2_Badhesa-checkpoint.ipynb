```python
# 1)

# a) 

import numpy as np
import matplotlib.pyplot as plt


# creating a list of numbers to plot for the continous signal
x=np.arange(0,9)                    
length1=len(x)
#print(length1)

# Creating data points for our continous signal to be compared to our sampled signal, values are from 0-8 with steps of 0.1
AS=np.arange(0,9,0.1)
length=len(AS)

# Plotting both the straight line of y=x and the step function using the plt.step to show the sample holding a particular 
# value while the original signal changes
plt.axis([0,8,0,8])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.title("'Rounding Down' Sampling of Analog Signal ")
plt.plot(x)
plt.step(x,x-1)
plt.show()

# a)

# Initializing arrays for the error calculations
QE=[0]*length
i=0
j=0
k=-9                         # Ensures that k is 0 at the beggining of the for loop, needed multiple variables to 
                             # subtract arrays with different lengths below:
for i in range(0,length1):
    k +=10
    j=0
    for j in range(0,length1):
        #print(AS[j+k],x[i],i,j,k)
        QE[j+k]=AS[j+k]-x[i]           # Subtract the sample value from the actual signal value
#print(QE)

# Plot the error to see how it varies
plt.axis([0,7,0,1.25])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.title("Error in Rounding Down Sampled Signal ")
plt.plot(AS,QE)
plt.show()

# b)

print("The maximum error is %.2f" % (max(QE)))

# c)

# Plotting a Signal and First half of the signal ranges to a sampled value
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.title("Mid Point Sampling of Analog Singal")
plt.plot(x)
plt.step(x+0.5,x)
plt.axis([0,8,0,8])
plt.show()

# Initializing array for above sampling method error
Q=[0]*length
i=0
h=0
for i in range(0,1):                     # The first five signal values are calculated for error seperately to avoid  
    for h in range(0,5):                 # cluster of confusion in remaining data analysis
        Q[h]=AS[h]-x[i]
        #print(AS[h],x[i],Q[h],h)
# print("here")
i=0
j=0
k=-5                        # Same method of subtraction using 3 variables except k will start from 5 to avoid 
for i in range(0,length1-1): # adding new values to first five data points
    k +=10
    j=0
    for j in range(0,length1+1):
        #print(AS[j+k],x[i+1],i,j,k)
        Q[j+k]=AS[j+k]-x[i+1]
        #print(Q[j+k])

        
# The last five points are calculated seperately to avoid confusion        
#print("here")
h=0       
for h in range(length-1,length-6,-1):
    Q[h]=AS[h]-9
    #print(Q[h],h)       
#print(length)
#print("here")
#print(Q)

plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.title("Error in Mid Point Sampled Signal ")
plt.plot(AS,Q)
plt.show()
    
# d)

print("The maximum error is %.3f" % (max(np.abs(Q))))  # maximum error is actually on the negative side so took the abs val
    
```

```python
#2 

# a)

# Generating a "continous" sinusoidal signal with the following parameters
A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=5 # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve


# Sampling the signal at a frequency 6 times the frequency of the continous signal(fs1)
A1=2 # Amplitude of the sin wave
f1=5.0 # frequency of the sin wave in Hz
T1=1/f1
fs1=6.0*f1
Ts1=1/fs1 # horizontal spacing of the points that "fake" a continuous signal
cycles1=5 # how many cycles of the wave to save
t1=np.arange(0,cycles1*T1,Ts1) # horizontal coordinate of your sin curve
x1=A1*np.sin(2*np.pi*f1*t1) # vertical value of your sin curve


# Plotting the continous signal as plt.plot and the sampled as spikes using plt.stem
plt.figure(figsize=(12,4))
plt.axis([0,1,-2,2]) # were not plotting all cycles so we can see whats going on
plt.stem(t1,x1,use_line_collection=True)
plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")
plt.show()

# b)

# Same as part a) except sampling at 12 hz (fs2)
A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=5 # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

A2=2 # Amplitude of the sin wave
f2=5.0 # frequency of the sin wave in Hz
T2=1/f2
fs2=f2*(12/5)
Ts2=1/fs2 # horizontal spacing of the points that "fake" a continuous signal
cycles2=5 # how many cycles of the wave to save
t2=np.arange(0,cycles2*T2,Ts2) # horizontal coordinate of your sin curve
x2=A2*np.sin(2*np.pi*f2*t2) # vertical value of your sin curve

plt.figure(figsize=(12,4))
plt.axis([0,1,-2,2]) # were not plotting all cycles so we can see whats going on
plt.stem(t2,x2,use_line_collection=True)
plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")
plt.show()


# c)

# Same as part a) except sampling at 10 hz (fs3)
A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=5 # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

A3=2 # Amplitude of the sin wave
f3=5.0 # frequency of the sin wave in Hz
T3=1/f3
fs3=f3*2
Ts3=1/fs3 # horizontal spacing of the points that "fake" a continuous signal
cycles3=5 # how many cycles of the wave to save
t3=np.arange(0,cycles3*T3,Ts3) # horizontal coordinate of your sin curve
x3=A3*np.sin(2*np.pi*f3*t3) # vertical value of your sin curve

plt.figure(figsize=(12,4))
plt.axis([0,1,-2,2]) # were not plotting all cycles so we can see whats going on
plt.stem(t3,x3,use_line_collection=True)
plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")
plt.show()

# d)

# Same as part a) except sampling at 7 hz (fs4)
A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=5 # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

A4=2 # Amplitude of the sin wave
f4=5.0 # frequency of the sin wave in Hz
T4=1/f4
fs4=f4*(7/5)
Ts4=1/fs4 # horizontal spacing of the points that "fake" a continuous signal
cycles4=5 # how many cycles of the wave to save
t4=np.arange(0,cycles4*T4,Ts4) # horizontal coordinate of your sin curve
x4=A4*np.sin(2*np.pi*f4*t4) # vertical value of your sin curve

plt.figure(figsize=(12,4))
plt.axis([0,1,-2,2]) # were not plotting all cycles so we can see whats going on
plt.stem(t4,x4,use_line_collection=True)
plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")
plt.show()
```

```python
# 3

# a)

# Fourier transforming the samples taken from Q2 to convert to frequency domain
# And plotting fft(xN) points to notice spikes at expected values of 5 and a conjugate frequency as well
fft1 = np.fft.fft(x1) #xsamp is your stem plot data
N1 = x1.size
Tsamp1=1/30
f1 = np.linspace(0, 1 / Tsamp1, N1)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Fourier Transform Plot")
plt.stem(f1,np.abs(fft1)/N1,use_line_collection=True)
plt.show()

fft2 = np.fft.fft(x2) #xsamp is your stem plot data
N2 = x2.size
Tsamp2=1/12
f2 = np.linspace(0, 1 / Tsamp2, N2)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Fourier Transform Plot")
plt.stem(f2,np.abs(fft2)/N2,use_line_collection=True)
plt.show()

fft3 = np.fft.fft(x3) #xsamp is your stem plot data
N3 = x3.size
fsamp3=10
Tsamp3=1/fsamp3
f3 = np.linspace(0, 1 / Tsamp3, N3)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Fourier Transform Plot")
plt.stem(f3,np.abs(fft3)/N3,use_line_collection=True)
plt.show()

fft4 = np.fft.fft(x4) #xsamp is your stem plot data
N4 = x4.size
Tsamp4=1/7
f4 = np.linspace(0, 1 / Tsamp4, N4)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Fourier Transform Plot")
plt.stem(f4,np.abs(fft4)/N4,use_line_collection=True)
plt.show()

#print(fft1,fft2,fft3,fft4)
```
