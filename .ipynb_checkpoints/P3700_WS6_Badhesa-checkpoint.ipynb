```python
import numpy as np
import matplotlib.pyplot as plt
import cmath


nsamp=100 # number of sampling in $\Omega$ space...faking a continuous function
Homega=[0.]*nsamp # Frequency response
ndow=7
hn=[1.]*ndow #impulse response of the filter (7 term square pulse)


ii=0 ## Counter for building the frequency response
while ii<nsamp:
    i=0
    omega=ii*np.pi/nsamp ## converting counter to a digital frequency (zero to pi)
    while i<ndow:
        Homega[ii]+=hn[i]*cmath.exp(complex(0, -(i*omega))) #building H at an Omega
        i+=1
    ii+=1

plt.plot(np.abs(Homega))
plt.show()

```

```python
# Using different types of window functions (Hanning,Hamming, Blackman), we are creating an array of windows to account
# for the begginning and ends of our filtered signals. THe functions take an input of total width of the window and
# is centred at the middle of the window and not zero.


def han(N):
    windw=[0.]*N
    for i in range(-int(N/2),int(N/2)):
        windw[i+int(N/2)]=0.5+0.5*np.cos(2*np.pi*i/(N-1))
    return windw

def ham(N):
    windw=[0.]*N
    for i in range(int(-N/2),int(N/2)):
        windw[i+int(N/2)]=0.54+0.46*np.cos(2*np.pi*i/(N-1))
    return windw
def BM(N):
    windw=[0.]*N
    for i in range(int(-N/2),int(N/2)):
        windw[i+int(N/2)]=0.5+0.5*np.cos(2*np.pi*i/(N-1))+0.08*np.cos(4*np.pi*i/(N-1))
    return windw
```

The cells below check the frequency response of our window arrays generated with the functions created above

```python
nsamp=100 # number of sampling in $\Omega$ space...faking a continuous function
Homega=[0.]*nsamp # Frequency response
ndow=35
hn=han(35) #impulse response of the filter (7 term square pulse)


ii=0 ## Counter for building the frequency response
while ii<nsamp:
    i=0
    omega=ii*np.pi/nsamp ## converting counter to a digital frequency (zero to pi)
    while i<ndow:
        Homega[ii]+=hn[i]*cmath.exp(complex(0, -(i*omega))) #building H at an Omega
        i+=1
    ii+=1

plt.plot(np.abs(Homega))
plt.show()
```
