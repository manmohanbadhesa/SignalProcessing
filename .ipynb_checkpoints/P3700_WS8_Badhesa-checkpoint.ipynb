```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

fs=1500
N=750
F=np.linspace(0,750,N)

Ho=[0.]*N
Hw=[0.]*N
Ho3=[0.]*N

for i in range(0,N):
    Ho[i]=(2000*cmath.exp(complex(0,-i*2*np.pi/fs))+2000)/(5000*cmath.exp(complex(0,-i*np.pi/fs))-1000)
    Hw[i]=1/np.sqrt((((2*np.pi*i)**2)/(2000**2))+1)
    Ho3[i]=(2360.42*cmath.exp(complex(0,-i*2*np.pi/fs))+2360.42)/(5360.42*cmath.exp(complex(0,-i*np.pi/fs))-639.58)
plt.figure(figsize=(12,8))
plt.plot(F,Ho, label = "H02")
plt.plot(F,Ho3, label = "H03")
plt.plot(F,Hw, label = "Hw")
plt.xlabel('\u03A9')
plt.ylabel('H(\u03A9)')
plt.title("Frequency Response")
plt.legend(loc= 'upper right')
plt.show()

```
