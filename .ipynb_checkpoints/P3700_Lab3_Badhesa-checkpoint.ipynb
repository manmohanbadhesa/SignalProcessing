# 1)

```python
import numpy as np

# Creating low-pass, high-pass, band-pass, and band-rejct recursive filters
# All filters simply take in the list and an appropriate variable input,
# calculate the variables for the filters using equations 19-2 to 19-5,
# 19-7,and 19-8 and then apply to the input list to get an output list
#  using equation 19-1(the Recursion equation)
def lowRecur(list1,fc):
    x=np.exp(-2*np.pi*fc)
    a0=1-x
    b1=x
    leng=len(list1)
    y=[0]*leng
    y[0]=a0*list1[0]
    i=1
    for i in range(1,leng):
        y[i]=a0*list1[i]+b1*y[i-1]
    return y

def highRecur(list1,fc):
    x=np.exp(-2*np.pi*fc)
    a0=(1+x)/2
    a1=-(1+x)/2
    b1=x
    leng=len(list1)
    y=[0]*leng
    y[0]=a0*list1[0]
    i=1
    for i in range(1,leng):
        y[i]=a0*list1[i]+a1*list1[i-1]+b1*y[i-1]
    return y

def BpRecur(list1,f,bw):
    R=1-(3*bw)
    K=(1-(2*R*np.cos(2*np.pi*f))+(R**2))/(2-(2*np.cos(2*np.pi*f)))
    a0=1-K
    a1=2*(K-R)*np.cos(2*np.pi*f)
    a2=(R**2)-K
    b1=2*R*np.cos(2*np.pi*f)
    b2=-(R**2)
    leng=len(list1)
    y=[0]*leng
    y[0]=a0*list1[0]
    y[1]=a0*list1[1]+a1*list1[0]+b1*y[0]
    i=2
    for i in range(2,leng):
        y[i]=a0*list1[i]+a1*list1[i-1]+a2*list1[i-2]+b1*y[i-1]+b2*y[i-2]
    return y

def BrRecur(list1,f,bw):
    R=1-(3*bw)
    K=(1-(2*R*np.cos(2*np.pi*f))+(R**2))/(2-(2*np.cos(2*np.pi*f)))
    a0=K
    a1=-2*K*np.cos(2*np.pi*f)
    a2=K
    b1=2*R*np.cos(2*np.pi*f)
    b2=-(R**2)
    leng=len(list1)
    y=[0]*leng
    y[0]=a0*list1[0]
    y[1]=a0*list1[1]+a1*list1[0]+b1*y[0]
    i=2
    for i in range(2,leng):
        y[i]=a0*list1[i]+a1*list1[i-1]+a2*list1[i-2]+b1*y[i-1]+b2*y[i-2]
    return y

    
```

# 2)

```python
from scipy import signal
import matplotlib.pyplot as plt

# Square wave for 0.01 seconds
f=500
T=0.01
fs=50000
t = np.linspace(0,T,int(T*fs), endpoint=False)
x=signal.square(2 * np.pi * f * t)
N=len(x)
noise=np.random.uniform(-0.3,0.3,N)

# adding noise to the square
noisyS=[0]*N
i=0
for i in range(0,N):
    noisyS[i]+=x[i]+noise[i]


plt.figure(figsize=(12,8))
plt.plot(t, noisyS)
plt.ylim(-2, 2)
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Noisy Square Wave")
plt.grid(True)
plt.show()
```

# 3a)

```python

# Applying the filter on the noisy square wave by just inputting the
# list to the functions created in Q1
fc1=0.12
fil1=lowRecur(noisyS,fc1)

plt.figure(figsize=(12,8))
plt.plot(t, fil1)
plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc1*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Low-Pass Recursive Filter Square Wave")
plt.grid(True)
plt.show()
```

# 3b)

```python

# Applying filter with cutoff frequency 0.05 which can be seen to begin to 
# distort the wave 
fc2=0.05
fil2=lowRecur(noisyS,fc2)

plt.figure(figsize=(12,8))
plt.plot(t, fil2)
plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc2*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Low-Pass Recursive Over-Filtering Square Wave")
plt.grid(True)
plt.show()
```

# 3c)

```python

# function to low-pass filter using equation 19-9 which depends on the
# output next in list rather than the previous output, which is why the 
# the filter calculates outputs by going backwards in the input list
def lowRecurForward(list1,fc):
    x=np.exp(-2*np.pi*fc)
    a0=1-x
    b1=x
    leng=len(list1)+1
    y=[0]*(leng)
    y[leng-1]=0
    i=1
    for i in range(1,leng):
        y[leng-i-1]=a0*list1[leng-i-1]+b1*y[leng-i]
        #print(i,leng-i-1,leng-i)
    del y[leng-1]
    return y

#h123=[1,4,7,8,5,7]              Shorter Array I created for testing
#o12=lowRecurForward(h123,0.2)

# applying the filter to an already filtered list but using the above 
# equation to make the output symmetric
fc3=0.05
filForward=lowRecurForward(fil1,fc3)


plt.figure(figsize=(12,8))
plt.plot(t, filForward)
plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc3*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Low-Pass Recursive Filter Forward and Backwards Square Wave")
plt.grid(True)
plt.show()
```

```python
# Lab 2 low pass moving average filter

# Creating the convolution function which convolves two lists into one 
# using a simple mathematical expression
def myconv(list1,list2):
    len1=len(list1)
    len2=len(list2)
    
    # Below we assign the longer list to x and shorter list to h to ensure 
    # the elements compiling efficiency
    if len1> len2:
        h=list2
        x=list1
    else:
        h=list1
        x=list2
    len3=len1+len2-1
    #print(len3)
    y=[0]*len3
    i=0
    # below we loop through summing up our elements as h moves through x 
    for i in range(0,len3):
        j=0
        for j in range(0,len(h)):
            if i-j >=0:
                if i-j < len(x):
                    y[i]+=x[i-j]*h[j]
                    #print(y[i], x[i-j], h[j], i,j,i-j)
    return y

def lowSquare(L1):
    lsquare=[0]*L1
    i = 0
    for i in range(0,L1):
        lsquare[i]=1/L1
    return lsquare



```

```python
# Running the filter

# Added the kernel generation as part of the timed part because that is
# also how the filter works so better to time that as well

time_start = time.process_time()
#run your code,  Just the filter part timed
kernels=lowSquare(11)
fil4=myconv(kernels,noisyS2)

time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)


plt.figure(figsize=(12,8))
plt.plot(t2[0:1000], fil4[0:1000])
plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc2*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Low-Pass Moving Average Filtered Square Wave")
plt.grid(True)
plt.show()

# The average run time was 0.531 for the moving average filter for 3 runs. 
# The time for the 31 point moving average is about 16 times slower than 
# the recursive filter 
```

```python
# Using the wave provided, chose 100Hz cutoff frequency, 0.04 in sampling
# frequency freq spectrum. 

A8=2
f8=50.0
T8=1/f8
fs8=50*f8
Ts8=1/fs8
cycles8=3
t8=np.arange(0,cycles8*T8,Ts8)
i=0
x8=[0.]*len(t8)
while i < len(t8):
    x8[i]=A8*np.sin(2*np.pi*f8*t8[i])+A8*np.sin(0.2*np.pi*f8*t8[i])+np.random.normal(0,0.2)
    i+=1

    
fc06=0.04
high01=highRecur(x8,fc06)

plt.figure(figsize=(12,8))
plt.plot(t8, x8)
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Unfiltered Sin Wave")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,8))
plt.plot(t8, high01)
plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc6*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("High-Pass Recursive Filtered Sin Wave")
plt.grid(True)
plt.show()

# The original wave was unstable and shifting up and the filter removed a 
# frequency which made it stable and not elevate upwards
```

```python
import IPython
from scipy.io import wavfile

# CFA from Lab2

IPython.display.Audio("CFA.wav")
fs012, data2 = wavfile.read('CFA.wav') #fs012 is the sampling rate
CFA = data2 # Left Channel


plt.figure(figsize=(24,8))
plt.plot(CFA)
plt.show()
IPython.display.Audio("CFA.wav")
```

```python
# Band Pass below on CFA to allow only A
A=440.01
bw1=0.001
bw2=0.01
cycles012=10
T012=1/fs012
t012=np.arange(0,len(CFA))

bPass=BpRecur(CFA,A,bw1)


# Band reject on CFA to remove A note
bReject=BrRecur(CFA,A,bw2)


plt.figure(figsize=(12,8))
plt.plot(t012[0:500], bPass[0:500])
#plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc6*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Band Pass Filtered Square Wave")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,8))
plt.plot(t012[0:500], bReject[0:500])
#plt.ylim(-2, 2)
plt.text(0,1.5, 'fc= %5.2f'  % (fc6*fs))
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.title("Band Reject Filtered Square Wave")
plt.grid(True)
plt.show()
```

```python
# Fourier Transform of Original and Band pass filtered wave, we can see
# the magnitude of the signal around C and F notes has decreased while the
# A note has also decreased but is still a significant magnitude compared
# to the C and F

fft2 = np.fft.fft(CFA) #xsamp is your stem plot data
N2 = len(CFA)
Tsamp2=1/fs012
f013 = np.linspace(0, 1 / Tsamp2, N2)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.figure(figsize=(24,8))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Before Fourier Transform Plot")
plt.stem(f013[0:500],np.abs(fft2)[0:500]/N2,use_line_collection=True)
plt.show()

fft1 = np.fft.fft(bPass) #xsamp is your stem plot data
N1 = len(bPass)
Tsamp1=1/fs012
f012 = np.linspace(0, 1 / Tsamp1, N1)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.figure(figsize=(24,8))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("After Band Pass Fourier Transform Plot")
plt.stem(f012[0:500],np.abs(fft1)[0:500]/N1,use_line_collection=True)
plt.show()
```

```python
# Rejecting the A frequency decreased the A note magnitude signifcantly 
# but didn't remove the signal entierly. It did reduce the C and F 
# notes but still were a significant signal just like the A in Band pass

fft2 = np.fft.fft(CFA) #xsamp is your stem plot data
N2 = len(CFA)
Tsamp2=1/fs012
f013 = np.linspace(0, 1 / Tsamp2, N2)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.figure(figsize=(24,8))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("Before Fourier Transform Plot")
plt.stem(f013[0:500],np.abs(fft2)[0:500]/N2,use_line_collection=True)
plt.show()

fft6 = np.fft.fft(bReject) #xsamp is your stem plot data
N6 = len(bReject)
Tsamp6=1/fs012
f016 = np.linspace(0, 1 / Tsamp6, N6)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.figure(figsize=(24,8))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.title("After Band Reject Fourier Transform Plot")
plt.stem(f016[0:500],np.abs(fft6)[0:500]/N6,use_line_collection=True)
plt.show()
```
