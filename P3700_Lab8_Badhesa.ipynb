```python
import numpy as np
import matplotlib.pyplot as plt

# Histogram equalization function
def histoE(list1,maxG):
    # N is the number of rows
    N=len(list1)
    # n is the number of columns, I may have them opposite but n is the number of elements in each "dimension"
    n=len(list1[0])
    totalPixels=N*n
    # arrary which stores number of times a pixel is less than the index value and upto maxG
    array=[0.]*maxG
    # the output array after histogram equalization
    output=[[0 for x in range(n)] for y in range(N)]
    for i in range(0,maxG):
        for j in range(0,N):
            j=0
            for k in range(0,n):
                if list1[j][k] <= i:
                    array[i]+=1
    # output array is just the number of times counted as a function of the input list and using equalization equation
    for i in range(0,N):
        j=0
        for j in range(0,n):
            output[i][j]= (array[list1[i][j]]/totalPixels)*maxG
    return output
```

# 1 b)

```python
import scipy.misc

import time
time_start = time.process_time()
#run your code

# Testing the histogram equalization on the ascent image and converted to outputed as gray scale image

ascent = scipy.misc.ascent()

test= histoE(ascent,256)

plt.gray()
plt.imshow(test)
plt.show()

time_elapsed = (time.process_time() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])
```

```python
import matplotlib.image as mpimg

# loading the image and converting rgb to gray

img1=mpimg.imread("400ran.png")

gray1=np.floor(rgb2gray(img1)*128+0.5)
plt.figure(figsize=(12,8))
plt.imshow(gray1,cmap=plt.get_cmap('gray'))
plt.show()
```

# 2 b)

```python
# Finding a "2" in the image 

y1=6
y2=18
x1=12
x2=20


template = np.copy(gray1[y1:y2, x1:x2])
template-=template.mean()

#plt.figure(figsize=(12,8))
plt.imshow(template,cmap=plt.get_cmap('gray'))
plt.show()
```

# 2 c)

```python
from scipy import signal

# running correlate with the "2" found in the image

correlate=signal.correlate2d(gray1,template,boundary='symm', mode='same')
```

# 2 d)

```python
# Plotting the 2D correlation and can see the brightest spots are on the 2's and less brighter spots are also visible on
# 7's

plt.figure(figsize=(20,20))
plt.imshow(correlate,cmap=plt.get_cmap('gray'))
#plt.xlabel('Pixel')
#plt.ylabel('Pixel')
plt.title("Correlation")
plt.show()
```

# 2 e)

```python
# Histogram of the 2D correlation

plt.hist(correlate.ravel(), bins='auto') # Quick histogram with settings on 'auto'
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Correlation Histogram')
plt.yscale('log')
plt.show()
```

# 2 f) and g)

```python
# After altering the values of the pixel values to 0 for the values above the determined threshold value in 2e), it is 
# evident that the 2's can be seen as white spots and rest is black

N=len(correlate)
n=len(correlate[0])

for i in range(0,N):
    j=0 
    for j in range(0,n):
        if correlate[i][j]<125000:
            correlate[i][j]=0


            
plt.figure(figsize=(20,20))
plt.imshow(correlate,cmap=plt.get_cmap('gray'))
#plt.xlabel('Pixel')
#plt.ylabel('Pixel')
plt.title("Correlation")
plt.show()
```

```python
dice0=mpimg.imread("dice_gray.png")
avocado0=mpimg.imread("avocado_gray.png")

dice=np.floor(rgb2gray(dice0)*128+0.5)
plt.figure(figsize=(12,8))
plt.imshow(dice,cmap=plt.get_cmap('gray'))
plt.show()

avocado=np.floor(rgb2gray(avocado0)*128+0.5)
plt.figure(figsize=(12,8))
plt.imshow(avocado,cmap=plt.get_cmap('gray'))
plt.show()
```

```python
# Used the histogram to see the varying pixel values in order to find a way to add the dice to the avocado, initially tried
# 60 pixel value but it became a combination with the black spots on the dice becoming the avocado and many more 
# combinations

plt.hist(dice.ravel(), bins='auto') # Quick histogram with settings on 'auto'
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of Dice Pixels')
plt.yscale('log')
plt.show()
```

```python
# Finally, what worked was using the white background in the dice image was made equal to the avocado pixels white the 
# dice white had pixel values a little lower than the white background

c,r=len(dice[0]),len(dice) #w for width and h for height
avoDice=[[0 for x in range(c)] for y in range(r)]

for i in range(0,r):
    j=0
    for j in range(0,c):
        if dice[i][j]> 120:
            avoDice[i][j]=avocado[i][j]
        else:
            avoDice[i][j]=dice[i][j]

plt.figure(figsize=(12,8))
plt.imshow(avoDice,cmap=plt.get_cmap('gray'))
plt.show()
```
